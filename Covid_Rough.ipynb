#### Now that we have our data, lets find out: 
#### the Y variable (statewise positive)
#### the various X variables to be regressed along with Y

```python
Y=[]
Yp=[]
X = {
    'rural':[],
    'urban':[],
    'area':[],
    'density':[],
    'GR':[],
}
for i in df_25.index:
    state_pos = df_25['Positive'][i]
    state_test = df_25['TotalSamples'][i]
    state_pop = list(df_population[df_population['State / Union Territory']==df_25['State'][i]]['Population'])[0]
    X['rural'].append(list(df_population[df_population['State / Union Territory']==df_25['State'][i]]['Rural population'])[0]*100/state_pop)
    X['urban'].append(list(df_population[df_population['State / Union Territory']==df_25['State'][i]]['Urban population'])[0]*100/state_pop)
    X['area'].append(int(list(df_population[df_population['State / Union Territory']==df_25['State'][i]]['Area'])[0].split('\xa0')[0].replace(',','')))
    X['density'].append(float(list(df_population[df_population['State / Union Territory']==df_25['State'][i]]['Density'])[0].split('/')[0].replace(',','')))
    X['GR'].append(list(df_population[df_population['State / Union Territory']==df_25['State'][i]]['Gender Ratio'])[0]/1000)
    Y.append(state_pos*100*1000/(state_test*state_pop))
    Yp.append(state_pos*100/state_test)
print(Y)

print(X)
```

```python
for i in X:
    print("For statewise attribute: ",i)
    plt.scatter(X[i],Yp)
    plt.show();
    print("_"*80)
```

```python
for i in X:
    print("For statewise attribute: ",i)
    plt.scatter(X[i],Y)
    plt.show();
    print("_"*80)
```

## Findings: 
> The graphs of Urban and Rural are mirror images, showing a positive corellation in urban vs positive. We can conclude that the virus, as of 26/05/2020, is more prevalent in Urban regions. A possible explanation to the above is that a lot of international travellers would rather be coming to urban regions than rural regions. Also, testing facilities are better and more in number in urban regions, increasing the chance of catching people who would test positive for COVID-19. 

>Larger the state, more the number of positive cases.

>The virus is not highly gender-specific, as the plots seen are scattered/flat, implying that the virus has affected men and women nearly the same.

>If a line was to be plotted for density vs % positive we will observe a steep line, implying that with a small rise in density, we will have a huge rise in number of positive cases. This shows the level of spread of the virus, how rapid it has been.

```python
df_26_d = df_indiv_india[df_indiv_india['Date']=='26/05/20']
df_26_d['State/UnionTerritory'][2313] = 'Dadra and Nagar Haveli and Daman and Diu'
```

```python
Y=[]
X = {
    'rural':[],
    'urban':[],
    'area':[],
    'density':[],
    'GR':[],
}
for i in df_26_d.index[:-1]:  
    state_deaths = df_26_d['Deaths'][i] 
    state_pop = list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Population'])[0]
    X['rural'].append(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Rural population'])[0]*100/state_pop)
    X['urban'].append(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Urban population'])[0]*100/state_pop)
    X['area'].append(int(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Area'])[0].split('\xa0')[0].replace(',','')))
    X['density'].append(float(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Density'])[0].split('/')[0].replace(',','')))
    X['GR'].append(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Gender Ratio'])[0]/1000)
    Y.append(state_deaths*1000/state_pop)
print(Y)

print(X)
```

```python
regress(X,Y)
```

## Findings
> Urban/Rural: there is a weak positive co-relation of deaths per thousand vs %urban population. Though weird, this says that if one has tested positive for COVID 19, and is in a state with larger urban population, the person is more likely to die, than in rural. But this is given that there are not many positive cases in rural regions, as per this dataset. This observation cannot be taken as truth.

>Area, like in %positive cases, shows positive relation with the data. This can be explained, as there are more positive cases in larger states, implying more deaths in larger states

>The regression model has given maximum weightage to density. We could infer that dense regions have had more deaths, and there is a sharp increase in number of deaths with a small inrease in density.

>The model gives near 0 weightage ot Gender Ratio, implying that males and females have an equal chance of survival/death once affected. This virus is neither gender specific nor does it harm a particular gender more

## Part 2
Ratio of Dead to Recovered

```python
Y=[]
X = {
    'rural':[],
    'urban':[],
    'area':[],
    'density':[],
    'GR':[],
}
for i in df_26_d.index[:-1]:  
    state_deaths = df_26_d['Deaths'][i]
    state_recovered = df_26_d['Cured'][i]
    state_pop = list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Population'])[0]
    X['rural'].append(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Rural population'])[0]*100/state_pop)
    X['urban'].append(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Urban population'])[0]*100/state_pop)
    X['area'].append(int(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Area'])[0].split('\xa0')[0].replace(',','')))
    X['density'].append(float(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Density'])[0].split('/')[0].replace(',','')))
    X['GR'].append(list(df_population[df_population['State / Union Territory']==df_26_d['State/UnionTerritory'][i]]['Gender Ratio'])[0]/1000)
    if state_recovered != 0:
        Y.append(state_deaths/state_recovered)
    else:
        Y.append(0.0)
print(Y)

print(X)
```

```python
regress(X,Y)
```

## Findings:
>rural/urban/GR - no contribution: no difference made if the person is in an urban/rural region, or is male/female. the ratio of dead to cured is the same

>area considerable contribution

>density still being the largest contributor (has max weightage)
